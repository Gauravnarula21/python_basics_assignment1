```python
# Describe the role of predefined keywo$ds in Python and provide examples of how they are used in a program.

'''In Python, predefined keywords are reserved words that have special meanings and purposes in the language. These keywords are integral to Python's syntax and structure, and they cannot be used as identifiers (e.g., variable names or function names).
Understanding these keywords is essential for writing correct and effective Python code.

Here's a brief description of some key Python keywords and examples of how they are used in programs:

Purpose: Used to define a function.
'''

# def: Used to define a function.
    
def greet(name):
    return f"Hello, {name}!"

# if, elif, else: Used for conditional statements to execute code based on certain conditions.

x = 10
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x is equal to 5")
else:
    print("x is less than 5")

# for, while:Used for looping through code. for is used for iterating over sequences, while while loops continue as long as a condition is true.

# Using for loop
for i in range(5):
    print(i)

# Using while loop
count = 0
while count < 5:
    print(count)
    count += 1
    
# import: Used to include the functionality of other modules or packages.

import math
print(math.sqrt(16))

#class:Used to define a class, which is a blueprint for creating objects.

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."

person1 = Person("gaurav", 30)
print(person1.greet())

#return: Used to exit a function and optionally pass back a value.

def add(a, b):
    return a + b

result = add(5, 3)
print(result)

# try, except, finally: Used for exception handling. try block contains code that might raise an exception, except block handles the exception, and finally block contains code that will run regardless of whether an exception occurred or not.

try:
    x = 1 / 0
except ZeroDivisionError:
    print("You can't divide by zero!")
finally:
    print("This will always be executed.")


```

```python
# compare and contrast mutable and immutable object in python with example

'''In Python, the concepts of mutable and immutable objects refer to whether an object's 
state or content can be modified after it is created. Understanding the differences between 
mutable and immutable objects is crucial for writing efficient and bug-free code.'''

#1 Mutable Objects: Mutable objects are those whose state or content can be changed after they are created.

# list:

my_list = [1, 2, 3]
my_list.append(4)  # Modifies the list in place
print(my_list)  # Output: [1, 2, 3, 4]

# Dictionaries:

my_dict = {'a': 1, 'b': 2}
my_dict['c'] = 3  # Adds a new key-value pair
print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 3}

# Sets:

my_set = {1, 2, 3}
my_set.add(4)  # Modifies the set in place
print(my_set)  # Output: {1, 2, 3, 4}

#2 Immutable Objects: Immutable objects are those whose state or content cannot be changed once they are created.

# Tuples:
my_tuple = (1, 2, 3)
#my_tuple[1] = 4  # This would raise an error: TypeError

#Strings:
my_string = "hello"
new_string = my_string.upper()  # Creates a new string
print(new_string)  # Output: HELLO

#Integers:
x = 5
x += 1  # Creates a new integer object and assigns it to x
print(x)  # Output: 6
```

```python
#  Discuss the diffe$ent types of operators in Python and provide examples of how they are used.

'''In Python, operators are symbols used to perform operations on variables and values. 
Python supports a wide range of operators, categorized into different types. 
Here's an overview of the different types of operators in Python, along with examples of how they are used:'''

#1. Arithmetic Operators:These operators are used to perform basic mathematical operations.

# Addition(+), Subtraction(-), Division(/), multiplication(*), Floor Division (//),Modulus (%),Exponentiation (**)

a=10
b=5
result = a + b 
print(result)
result = a - b 
print(result)
result = a / b 
print(result)
result = a * b 
print(result)
result = a // b 
print(result)
result = a % b 
print(result)
result = a ** b 
print(result)

#2. Comparison Operators: These operators are used to compare values and return Boolean results (True or False).

# Equal to (==), Not equal to (!=), Greater than (>), Less than (<),Greater than or equal to (>=),Less than or equal to (<=)

a = 10
b = 5
result = a == b 
print(result)
result = a != b 
print(result)
result = a > b  
print(result)
result = a < b  
print(result)
result = a >= b  
print(result)
result = a <= b 
print(result)


#3. Logical Operators:These operators are used to perform logical operations on Boolean values.

# And (and), Or (or), Not (not),

a = True
b = False
result = a and b  
print(result)
result = a or b
print(result)
result = not a
print(result)

#4. Assignment Operators: These operators are used to assign values to variables.

# Assignment (=), Addition assignment (+=), Subtraction assignment (-=), Multiplication assignment (*=), Division assignment (/=),
# Floor Division assignment (//=), Modulus assignment (%=), Exponentiation assignment (**=)

a = 10  # assigns 10 to a
a += 5  # equivalent to a = a + 5, a is now 15
a -= 5  # equivalent to a = a - 5, a is now 5
a *= 5  # equivalent to a = a * 5, a is now 50
a /= 5  # equivalent to a = a / 5, a is now 2.0
a //= 3  # equivalent to a = a // 3, a is now 3
a %= 3  # equivalent to a = a % 3, a is now 1
a **= 3  # equivalent to a = a ** 3, a is now 1000

#5. Bitwise Operators: These operators perform operations on the binary representation of integers.

# Bitwise AND (&),Bitwise OR (|),Bitwise XOR (^),Bitwise NOT (~),Bitwise Left Shift (<<),Bitwise Right Shift (>>)

a = 10  # binary: 1010
b = 4   # binary: 0100
result = a & b  # result is 0 (binary: 0000)
print(result)
result = a | b  # result is 14 (binary: 1110)
print(result)
result = a ^ b  # result is 14 (binary: 1110)
print(result)
result = ~a  # result is -11 (binary: ...11110101)
print(result)
result = a << 2  # result is 40 (binary: 101000)
print(result)
result = a >> 2  # result is 2 (binary: 10)
print(result)

#6. Identity Operators: These operators compare the memory locations of two objects.
#Is (is):
a = [1, 2, 3]
b = a
result = (a is b)  # result is True
print(result)

#Is not (is not):
a = [1, 2, 3]
b = [1, 2, 3]
result = (a is not b)  # result is True (different objects)
print(result)

#7. Membership Operators:These operators test for membership in sequences (like strings, lists, or tuples).
#In (in):
a = 3
b = [1, 2, 3, 4]
result = (a in b)  # result is True
print(result)

#Not in (not in):
a = 5
b = [1, 2, 3, 4]
result = (a not in b)  # result is True
print(result)

```

```python
# Explain the concept of type casting in Python with examples.

'''Type casting in Python refers to the process of converting a variable from one data type to another.
This is often necessary when you need to perform operations between different types of data or 
when you need to ensure data is in a specific format. Python provides several built-in functions to perform 
type casting, such as int(), float(), str(), list(), tuple(), and dict().'''

#1.Implicit Type Conversion
# Python automatically converts 
# a to int 
a = 7
print(type(a)) 
 
# Python automatically converts 
# b to float 
b = 3.0
print(type(b)) 
 
# Python automatically converts 
# c to float as it is a float addition 
c = a + b 
print(c) 
print(type(c))
 
# Python automatically converts 
# d to float as it is a float multiplication
d = a * b
print(d)
print(type(d))

#2. Explicit Type Conversion 

# int variable
a = 5
 
# typecast to float
n = float(a)
 
print(n)
print(type(n))

# int variable
a = 5.9
 
# typecast to int
n = int(a)
 
print(n)
print(type(n))

# string variable
a = "5"
b = 't'
 
# typecast to int
n = int(a) 
 
print(n)
print(type(n))
 
print(int(b)) # Here, we are Converting string to int datatype in Python with int() function. If the given string is not number, then it will throw an error.
print(type(b))
```

```python
#How do conditional statements work in Python? Illust$ate with examples.

'''Conditional statements in Python are used to execute different blocks of code based on certain conditions.
They allow your program to make decisions and execute specific code paths depending on the truthiness of the conditions.'''

#if Statement: The if statement evaluates a condition (an expression that returns True or False). 
#If the condition is True, the code block inside the if statement is executed.

x = 10

if x > 5:
    print("x is greater than 5")

#elif Statement: Short for "else if," the elif statement allows you to check multiple conditions if the initial if condition is False. 
#You can have multiple elif statements to test different conditions

x = 10

if x > 15:
    print("x is greater than 15")
elif x > 5:
    print("x is greater than 5 but less than or equal to 15")
else:
    print("x is 5 or less")

#else Statement: The else statement is executed if all preceding if and elif conditions are False. 
#It is optional and does not require a condition.

x = 3

if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")

#Nested Conditional Statements

x = 10
y = 20

if x > 5:
    if y > 15:
        print("x is greater than 5 and y is greater than 15")
    else:
        print("x is greater than 5 but y is 15 or less")
else:
    print("x is 5 or less")

#Conditional Expressions (Ternary Operator):
#Python also supports a shorthand way to write conditionals using the conditional expression, also known as the ternary operator.

x = 10
message = "x is greater than 5" if x > 5 else "x is 5 or less"
print(message)

```

```python
#Desc$ibe the different types of loops in Python and their use cases with examples.

''' Python, loops are used to execute a block of code repeatedly based on a condition or over a sequence of elements. 
Python provides two primary types of loops:'''

#1. for Loop: The for loop is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of code for each item in the sequence. 
#It is generally used when you know the number of iterations in advance or when you are working with iterable objects.

# syntax: for variable in sequence:

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
    
    
for i in range(5):
    print(i)

word = "hello"
for char in word:
    print(char)

#2. while Loop:The while loop repeatedly executes a block of code as long as a given condition evaluates to True. 
#It is used when the number of iterations is not known beforehand and depends on a condition.

#syntax: while condition:

count = 0
while count < 5:
    print(count)
    count += 1

x = 10
while x > 0:
    print(x)
    x -= 2
    
# Python also provides several statements to control the flow of loops:

#break: Exits the loop prematurely.
for i in range(10):
    if i == 5:
        break
    print(i)

#continue: Skips the rest of the code inside the loop for the current iteration and proceeds to the next iteration
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
    
#else in Loops: An else block can be attached to both for and while loops.
#It is executed when the loop completes normally (i.e., not terminated by break).

for i in range(5):
    print(i)
else:
    print("Loop finished")

    
```
